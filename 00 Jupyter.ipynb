## Jupyter

Jupyter is a beautiful mix of markdown and code.

A Jupyter notebook is a sequence of cells. Cells may be of type `code` or `markdown`.

#### Jupyter Cells

Yes, this is a **markdown** cell. Cells may be of different type: code or markdown. Code may be python (as we are using a python kernel), but kernels for other languages exist as well, e.g. SQL, Java, etc. 

Also note that there is neither an In[.] nor an Out[.] on the left side of a markdown cell. You can switch the type of a cell through the menu *or* by using the shortcut ESC+M to switch to markdown and ESC+Y to switch to code.

```python
import altair as alt
from vega_datasets import data
import pandas as pd
source = data.cars()

alt.Chart(source).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Origin', 'Horsepower', 'Miles_per_Gallon']
).interactive()
```

```python
alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='Origin:N'
).properties(
    width=150,
    height=150
).repeat(
    row=['Horsepower', 'Acceleration', 'Miles_per_Gallon'],
    column=['Miles_per_Gallon', 'Acceleration', 'Horsepower']
).interactive()
```

Program state is accumulated in a notebook through every execution of a code cell. This has pros and cons. Sometimes you edit a cell but forget to execute it and wonder why you do not receive the expected result further down in your notebook. Sometimes you execute a cell but forgot to execute the cells above. 

**When in doubt:** go to Menu "Kernel" and hit "Restart&Run all". If the bug is still there, it is your fault .
